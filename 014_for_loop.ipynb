## **For Loop**

**range function**
***
> - range(included, excluded)
> - range(start, end, step)

```python
range(1,11)
```

```python
list(range(1,11))
```

```python
tuple(range(0,5))
```

```python
list(range(1,11,2))
```

```python
list(range(10,0,-1))
```

**Sequence**
> - Maintain order
___

```python
["siyam","hossain","sh"]
```

```python
("siyam","hossain","sh")
```

**Range based for loop**
___

```python
for i in range(1,11):
    print(i,end=" ")
```

```python
for i in range(1,11,2):
    print(i,end=" ")
```

**Sequence based for loop**
***

```python
for char in "siyam hossain":
    print(char, end=" ")
```

```python
for item in ["siyam", "hossain", "sh"]:
    print(item, end=" ")
```

```python
for item in (1,2,3,4,5):
    print(item, end=" ")

print()
for item in {5,6,8,7}:
    print(item, end=" ")
```
