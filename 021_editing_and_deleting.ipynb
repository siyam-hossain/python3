## **Editing and Deleting in Strings**

```python
c = "Hello"
print(c)
```

```python
# c[0] = 'X'

# not possible: cause strings are a Immutable Data type: can't change
```

**Strings are Immutable Type**

+ Mean strings are not modifiable, where as reassignmet is possible   
+ Therefore, editing is not possible in string

```python
# Immutable
# c[0] = "X"
```

```python
# Reassignment
c = "World"
print(c)
```

```python
# can't add item at the end
# c[5] = "X"
```

**Deletion**
+ also we can't delete specific item from string, in order to preventing mutation

```python
# print(c)
del c
# print(c)
```

```python
# Not possible: to delete specific item form a string
# It create mutability on that string

del c[0] 
```

```python
del c[:3:2]
```

---
+ Can reassign fully
+ Can delete string entirely
+ Can't delete string partially 
+ Can't add, modify string partially
