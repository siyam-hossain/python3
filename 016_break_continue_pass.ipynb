## **break**
useful for linear search

```python
for i in range(1,11):
    if i == 5:
        break
    print(i,end=" ")
```

## **continue**
out of stock skip

```python
for i in range(1,11):
    if i == 5:
        continue 
        # if condition meet then skip rest of the statement
        # for this this case only 
        # it skip 5, 
    print(i,end=" ")
```

## **pass**

```python
for i in range(1,11):
    pass
```
