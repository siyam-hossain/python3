### Operators
###### Operators are used to perform operations on variables and values. Python has the following operators:

> - Arithmetic operators
> - Comparison operators
> - Logical operators
> - Bitwise operators
> - Assignment operators
> - Identity operators
> - Membership operators


**Arithmetic Operators**

```python
x = 5
y = 2

print("x+y  : ",x+y)   # addition
print("x-y  : ",x-y)   # subtraction
print("x*y  : ",x*y)   # multiplication
print("x/y  : ",x/y)   # division : true division
print("x%y  : ",x%y)   # modulo or remainder calculation
print("x^y  : ",x**y)  # power : x^y
print("x//y : ",x//y)  # integer division 
```

**Comparison Operators**

```python
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
print(x != y)
print(x == y)
```

**Logical Operators**

```python
a = True
b = False

print(a or b)
print(a and b)
print(not a)
print(not b)
```

**Bitwise Operators**
###### Perform on binary value. Example: Image processing filter

```python
i = 2
j = 3

# i = 2 = 010
# j = 3 = 011
#--------------
# &     = 010 ---> 2
#------------------
# |     = 011 ---> 3

print(i & j) 
print(i | j)
```

```python
# i / 2 = 2/2 = 1
# i * 2 = 1/2 = 0

# never store true division value 
print(i >> 2)
```

```python
# i * 2 = 2*2 = 4
# i * 2 = 4*2 = 8
# i * 2 = 8*2 = 16

print(i << 3)
```

```python
# complement
print(~i)
```

**Assignment Operator**

```python
c = 3
print(c)

# c = c+3
c += 3
print(c)

c -= 2
print(c)

c *= 4
print(c)

c /= 2
print(c)

# Invalid
# c++
# ++c
```

**Identity Operator**
###### is operator define whether (left is right) their memory location same or not

```python
d = 2
s = 2
print(d is s)

d = "hello"
s = "hello"
print(d is s)

d = "hello-world"
s = "hello-world"
print(d is s)

d = [1,2,3,4]
s = [1,2,3,4]
print(d is s)
```

**Membership Operator**
###### can use in : string, tuple, list, dictionary

```python
x = "Siyam Hossain" 

print("S" in x)
print("E" in x)
```

```python
x = [1,2,3,4]

print(3 in x)
print(5 in x)
```
