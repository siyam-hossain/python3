**Problem**

+ list
+ input: [1,1,1,2,2,3,4,4]
+ output: [1,2,3,4]

```python
l1 = [1,1,1,2,2,3,4,4]
l2 = [1,2,3,1]

empty = []

for i in l1:
    if i not in empty:
        empty.append(i)

print(empty)

empty.clear()
for i in l2:
    if i not in empty:
        empty.append(i)
print(empty)
```
