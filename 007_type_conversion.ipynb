### **Type conversion**

**implict**
> - automatically type conversion performed

**explict**
> - we the human define when to type convert

```python
# implicit
4 + 5.5 #float = int + float
```

```python
5 + 6 + 65j
```

```python
4.5 + 5 + 65j
```

```python
# Explicit
firstNum = input("Enter 1st Number: ")
secondNum = input("Enter 2nd Number: ")

sum = firstNum+secondNum # instead of addition it perform concatanation
print("firstNum + secondNum = ",sum)
```

```python
firstNum = int(input("Enter 1st Number: "))
secondNum = int(input("Enter 2nd Number: "))

sum = firstNum+secondNum
print("firstNum + secondNum = ",sum)
```
