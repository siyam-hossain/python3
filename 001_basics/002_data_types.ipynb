# Data type
# -> Basic (integer, float, complex, boolean and string)
# -> Container types (List, Tuples, Sets and Dictionary)
# -> User-defined types (class)

# Basic type

```python
# -> integer
print(4)
print(100000000000000000000000000000000000000000000)
# max range
print(1e308)
# range exit
print(1e309)
```

```python
# float
print(4.5)
print(1.7e308)
print(1.7e309)
```

```python
# boolean
print(True)
print(False)
```

# complex

```python
print(4+5j)
```

```python
# string
print('Bangladesh')
print("Bangladesh")
print("""Bangladesh""")
```

# container

```python
# list
print([1,2,3,4,5])
```

```python
# tuple
print((1,2,3,4,5))
```

```python
# sets
print({1,2,3,4,5})
```

```python
# dict
print({"name":"siyam", "Age":30, "gender":"Male" })
```
