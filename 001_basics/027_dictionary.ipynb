> ## **Dictionary** 

+ key : value      pair
+ separated by , 
+ Dictionary has no indexing
+ Dictionary is a mutable data types
+ keys --> must immutable,
+ values --> they can be mutable
+ keys should be unique

+ mutable : list/ sets/ dictionary
+ immutable : strings/  tuples/  int/ float/ boolean/ complex

> #### **Create**

```python
d = {}
d
```

```python
type(d)
```

```python
d = dict({"name":"siyam"})
d
```

```python
d = {"name":"siyam", "gender": "male"}
d
```

> #### **keys --> must immutable**

```python
d1 = {[1,2,3]: "siyam"}
d1
```

```python
d1 = {(1,2,3): "siyam"}
d1
```

> #### **keys should be unique**

```python
d2 = {"name":"siyam", "name":"hossain", "gender": "male"}
d2
```

> #### **2D dictionary**

```python
d3 = {"name":"siyam", "college":"dmrc", "marks":{"math":95,"eee":93}}
d3
```
