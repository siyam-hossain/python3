## **Accessing Substrings from a string**

**Concept of Indexing**

```python
c = "hello"
print(c)
```

```python
print(c[0])
```

```python
for i in range(0,len(c)):
    print(c[i],end=" ")
```

**Types of Indexing**

can access one character at a time

+ Positive Indexing 
    > - [start from left --> right]
+ Negative Indexing 
    > - [start from right --> left]


```python
# positive
print(c[0])

# negative
print(c[-1])
```

**Slicing**

can access a range of character
- string [ start : end ]
- end not included
---
+ string [ start : end : step ]

```python
c = "siyam hossain"
print(c)
```

```python
print(c[0:5])
```

```python
print(c[5:13])
```

```python
# string[start : all]
print(c[2:])
print(c[0:])
```

```python
# string[all : end] , end excluded
print(c[:13])
print(c[:5])
```

```python
# string[begin : end]
print(c[:])
```

```python
# with step
print(c[0:13:2])
print(c[0:5:3])
print(c[5:13:3])
```

```python
# with negative 
print(c[-1 : -13 : -1])
print(c[-13 : -1 : 1])
print(c[-13 :  : 1])
print(c[ :  : -1])
```
