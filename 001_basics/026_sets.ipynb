> #### **Edit not possible**

```python
s1[2] = 1000
s1
```

```python
# s1
print(s1)
# s1 memory address
print(id(s1))
```

```python
list = list(s1)
list
```

```python
list[1] = 2000
list
```

```python
print("before: ",s1)
print("before address: ",id(s1))

s1 = set(list)
print("after: ",s1)
print("after address: ",id(s1))
# therefore we can say, since memory address is different s1 not edited
```
