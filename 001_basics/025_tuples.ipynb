## **Tuples**

> tuples are immutable data-structure

> tuples are read-only data type

> when data integrity is importent

+ create
+ access
+ edit
+ add
+ delete
+ operations
+ functions

**Create**

```python
t1 = ()
t1
```

```python
# homogeneous tuple
t2 = (1,2,3,4)
t2
```

```python
# hetarogeneous tuple
t3 = (1,3,4,"hi there", False, True, 3.1416, 4+5j)
t3
```

```python
t4 = (1,2,3,(4,5))
t4
```

**Exception case**

```python
t5 = (1)
print(t5)
print(type(t5))
```

```python
t5 = ("hello world")
print(t5)
print(type(t5))
```

for single item tuple

```python
t5 = (1,)
print(t5)
print(type(t5))
```

```python
t6 = tuple()
t6
```

```python
t6 = tuple("hello")
t6
```

```python
t6 = tuple((1,2,3,4))
t6
```

```python
t6 = tuple([1,2,3,4])
t6
```
