## **if, else, elif, nested if else elif**

```python
# email : siyam@gmail.com
# password: 1234

email = input("Enter email: ")

if '@' in email:
    password = input("Enter password")
    
    if(email == "siyam@gmail.com" and password == "1234"):
        print("Welcome to SH")

    # partially right
    elif(email=="siyam@gmail.com" and password != "1234"):
        print("Incorrect password")
        password = input("Enter password")
        
        if(password == "1234"):
            print("Welcome to SH")
        else:
            print("Again wrong password")
    else:
        print("Wrong credentials")
    
else:
    print("Wrong email pattern")
```
