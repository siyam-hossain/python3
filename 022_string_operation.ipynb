**Loops on strings**

```python
name = "siyam hossain"
for ch in name:
    print(ch, end=" ")
```

```python
name = "siyam hossain"
for ch in name[1:11:2]:
    print(ch, end=" ")
```

```python
name = "siyam hossain"
for ch in name[::-1]:
    print(ch, end=" ")
```
