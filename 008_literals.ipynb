**Literals**
---
Literal is a row data given in a variable. In python, there are various types of literals they are as follows:
> + Numeric Literals
> + String Literals
> + Boolean Literals
> + Special Literals

>**1. Numeric Literals**

```python
a = 0b1010 #binary literals
b = 1000 # decimal literals
c = 0o310 # octal literals
d = 0x12c # hexadecimal literals


# float literal
f1 = 45.5
f2 = 1.6e2 # 1.6*(10^2)
f3 = 1.5e-3 # 1.5*(10^-3)

# complex literal
x = 3.1416j

print(a,b,c,d)
print(f1,f2,f3)
print(x, x.imag, x.real)
```

>**2. String Literals**

```python
string = 'This is python'
strings = "this is python"
char = "s"
multiline_str = """This is a multiline string with more than one line code."""
unicode = u"\U0001F600\U0001F606\U0001F923"
raw_str = r"raw \n string"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode)
print(raw_str)
```

>**3. Boolean Literal**

```python
a = True + 4
b = False + 10

print("a: ",a)
print("b: ",b)
```

>**4. Special Literal**

```python
a = None
print(a)
```
